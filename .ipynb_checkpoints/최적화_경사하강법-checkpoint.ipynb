### Batch-size
- large batch -> Sharp minimizer
- small batch -> flat minimizer



- Sharp minimizer : Training Funtion과 Test Funtion 사이의 차이가 큼 (Generalization이 좋지 않음)
- flat minimizer : Training Funtion과 Test Funtion 사이의 차이가 작음 (Generalization이 좋음)


# Gradient descent
- 경사하강법
- 학습률을 정하는게 중요


# Momentum
- Gradient의 방향이 변하지 않도록
- momentum 과 현재의 Gradiant를 합친 accumulation을 활용해서 Gradiant를 update


# Adagrad
- 많이 변한 파라미터에 대해서는 적은 변화를, 반대의 경우에는 큰 변화를 줌(G)



- G가 계속 커지기 때문에 W가 업데이트 되지 않는 현상이 발생할 수 있음(학습이 안되는 문제)

# Adadelta
- Adagrad의 학습이 멈추는 문제를 막음
- learning rate가 없음
- 많이 사용하지 않음


# RMSprop
- Adagrad와 비슷한 방식
- gradient squares를 구할 때 exp moving average(이동평균)을 더해줌


# Adam
- 가장 많이 사용
- RMSprop + Momentum
