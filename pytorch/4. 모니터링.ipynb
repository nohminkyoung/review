# 모니터링

## tensorboard
- tensorflow의 프로젝트로 만들어진 시각화 도구
- 학습그래프, metric, 학습결과의 시각화 지원
- pytorch도 연결 가능

- scalar : metric등 상수 값의 연속을 표시
- graph : 모델의 computational graph표시 
- histogram : weight의 값의 분포를 표현
- image&text : 예측값고가 실제값을 비교 

여는 법 (jupyter상에서 수행)
1. %load_ext tensorboard 
2. %tensorboard --logdir{파일위치}
  

- add_000 : 000자리에 확인하고싶은 내용들을 작성
- write를 완료했으면 close나 flush를 해줘야함
