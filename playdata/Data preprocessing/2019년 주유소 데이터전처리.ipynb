```python
import json
# 경기도 시군구 경계 정보를 가진 geo-json 파일 불러오기
geo_path = './skorea_municipalities_geo_simple.json'
try:
    geo_data2 = json.load(open(geo_path, encoding='utf-8'))
except:
    geo_data2 = json.load(open(geo_path, encoding='utf-8-sig'))
```

```python
g_map = folium.Map(location=[37.5502,126.982], 
                   tiles='Stamen Terrain', zoom_start=9)
```

```python
data = gas_1231.groupby(['지역'])[['고급휘발유', '휘발유', '경유']].mean()[['휘발유']]
```

```python
data.head(1)
```

```python
data.index = [x.split()[-1] for x in data.index]
data['휘발유'] = data['휘발유'].astype(int)
```

```python
# Choropleth 클래스로 단계구분도 표시하기
folium.Choropleth(geo_data=geo_data2,    # 지도 경계
                 data =data,      # 표시하려는 데이터
                 columns = [data.index, '휘발유'],  # 열 지정 
                fill_color='YlOrRd', fill_opacity=0.7, line_opacity=0.3, 
               key_on = 'feature.properties.name').add_to(g_map)
```
