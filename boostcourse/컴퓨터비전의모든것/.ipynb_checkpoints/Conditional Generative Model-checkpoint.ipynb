# Conditional Generative Model

예시 
1. 가방의 스케치가 주어졌을 때 실제 가방의 이미지로 변환하는 task를 수행한다고 가정
2. 이러한 종류의 task를 image translation이라 하는데, 이렇게 입력과 출력 이미지의 도메인이 다를 때, 마치 언어를 번역하는 것과 유사
3. 이때 sketch of bag이라는 condition이 주어지고, 이 condition에 따라 이미지 X를 생성하는 형태의 task를 수행하는 모델을 conditional generative model이라고 함


응용사례
- 저음질의 오디오를 고음질로 변환하는 audio super resolution
- 중국어를 영어로 번역하는 machine translation
- 제목이 주어졌을 때 글을 작성하는 article generation 등

#### GAN
- 위조지폐범(Generator) : 이미지를 생성하는 역할
- 경찰(Discriminator) : 주어진 진짜와 가짜를 구별하는 역할
- Generator가 학습을 할 때 Discriminator이 검출을 하지 못하게 적대적으로 학습
- Discriminator는 가짜를 검출하기 위해 적대적으로 학습
- 상호작용으로 둘의 성능이 계속 올라감 => 적대적 학습법

#### GAN의 구조

- 랜덤코드 z를 Generator에 입력해 가짜 데이터를 생성 -> Discriminator로 판별

### Conditonal GAN

#### 구조

- conditional 정보를 넣어주기 위해 c라는 인풋을 추가

#### 응용사례
- 유명한 화가의 그림과 같은 스타일로 변환해주는 image-to-image translation
- 저해상도 이미지를 고해상도 이미지로 변환해주는 super resolution
- 오래된 흑백 사진을 컬러 사진으로 만들어주는 colorization 등

### super resolution

- 저해상도 이미지를 고해상도 이미지로 변환
- 대표적인 Conditonal GAN

- 이전에는 CNN구조를 활용해서 학습
    - MAE(L1) 혹은 MSE(L2) Loss만으로 naive한 regression model을 구현
- naive regression model을 활용하면, 해상도는 높아지지만 흐릿한 이미지를 얻게 되는 것을 확인
    - 픽셀 자체의 intensity 차를 loss로 설정하기 때문에, generator가 적당한 평균적인 픽셀 값으로 이미지를 생성하도록 학습되기 때문
- GAN의 경우 discriminator가 그러한 트릭을 통해 생성된 이미지를 대부분 가짜로 판별해내기 때문에, 이러한 문제를 완화
