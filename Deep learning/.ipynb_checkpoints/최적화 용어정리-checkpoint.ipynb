# Generalization
- 일반화 성능
- Generalization gap : train_error와 test_error의 차이
- Generalization이 좋다 = 학습과 테스트 사이의 성능차이가 크지 않다
- Generalization이 작다고 해서 무조건 성능이 좋은 모델은 아님(train_error 자체가 클 수 있음)


# Underfitting vs Overfitting
- Overfitting : 학습 데이터에 대해서는 성능이 좋지만 테스트 데이터에 대해선 성능이 좋지 않음
- Underfitting : 학습 데이터에 대한 성능도 좋지 않은 경우


# Cross-validation(=k-fold validation) 
- validation : 학습데이터의 일부로 만들어진 검증데이터
- 학습데이터를 k개로 나눈 후 k 개를 학습으로, 1개를 검증으로 해서 학습하는 행위를 k번 반복
- 최적의 하이퍼파라미터를 찾기 위해서 사용
- 최종의 학습에는 전체 train데이터를 활용해서 학습시킴


# Bias and Variance
- Variance : 정밀도(일관적인 출력을 보여주는지)
- Variance가 크면 같은 입력에 대해서 다른 출력들을 보여줌 => overfitting 가능성이 크다
- Bias : 정확도(평균적으로 답에 얼마나 가까운지)
- Bias가 크면 target 값과 멀다


# Bootstrapping
- 학습데이터 중 몇개만 추출해서 여러개의 모델을 만들어냄
- 하나의 입력에 대한 모든 모델의 출력값을 구함
- 출력값들이 얼마나 일치를 이루는지 확인 함으로써 모델의 불확실성을 예측

# Bagging 
-  Bootstrapping을 사용해서 어러개의 모델을 만들어냄
- 만들어진 여러개의 모델의 결과를 종합해 평균을 냄

# Boosting
- 여러개의 모델을 만들어 합친 후 하나의 모델을 만듦
- 여러개의 약한 모델로 강한 모델을 만들기
