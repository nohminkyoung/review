# Early Stopping
- validation data를 활용해서 overfitting 되는 시점 이전에 학습을 멈춤


# Parameter Norm Penalty
- 파라미터가 너무 커지지 않도록
- 파라미터값 전체를 제곱 후 합쳐 이 전체적인 값의 크기를 줄임
- 함수의 공간속에서 함수를 부드러운 함수로 보기 위함


# Data Augmentation
- 데이터 증강
- 기존의 데이터를 변환해 새로운 데이터로 이식
- 각도, 색깔, 뒤집기 등



# Noise Robustness
- 입력 data에 noise를 집어넣음
- Data Augmentation과의 차이 : 입력뿐 아니라 neural net의 weight에도 집어넣음


# Label smootion
- Data Augmentation과 비슷
- 데이터 두개를 뽑아서 두개를 섞어줌


# Batch Normalization
- 간단한 분류문제를 해결할 때 성능을 올려줌
- 활성화함수의 활성화값 또는 출력값을 정규화
- 신경망의 각 layer에서 데이터(배치)의 분포를 정규화하는 작업
